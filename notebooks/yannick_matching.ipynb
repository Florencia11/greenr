### Repeat for ingredients (simpler)

```python
# Make df

ingr_df = pd.DataFrame(ingr_list, columns = ['ingredient'])
ingr_df
```

```python
# Grab summaries

ingr_summaries = []
ingr_links = []

for i, query in enumerate(ingr_df['ingredient']):
    
    if wiki.page(query).exists():
        page_py = wiki.page(query)
                
    else:
        query = query
        querlist = wikipedia.search(query)
        try:
            page_py = wiki.page(querlist[0])
        except IndexError:
            pass
            
    try:
        ingr_summaries.append(page_py.summary)
        ingr_links.append(page_py.links.keys())
    except:
        ingr_summaries.append('nan')
        ingr_links.append('nan')
        
    if i % 100 == 0:
        print(i)
```

```python
from multiprocessing import Pool

def scrape(series):
    
    ingr_summaries = []
    ingr_links = []
        
    for i, query in enumerate(series):
    
        if wiki.page(query).exists():
            page_py = wiki.page(query)

        else:
            query = query
            querlist = wikipedia.search(query)
            try:
                page_py = wiki.page(querlist[0])
            except IndexError:
                pass

        try:
            ingr_summaries.append(page_py.summary)
            ingr_links.append(page_py.links.keys())
        except:
            ingr_summaries.append('nan')
            ingr_links.append('nan')

        if i % 100 == 0:
            print(i)
            
    return ingr_summaries, ingr_links

p = Pool(20)

p.map(scrape, ingr_df['ingredient'])
    
```

### Tokenization (etc.) of summaries and links

```python
import string

import nltk
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
```

```python
df = df_cats_ingrs.copy()
df = df.reset_index(drop=True).dropna()
```

```python
# Dictkeys to list

df['links'] = [list(linkset) for linkset in df['links']]
```

```python
# Remove punctuation

for punctuation in string.punctuation:
    df['summaries'] = [str(text).replace(punctuation, '') for text in df['summaries']]
    df['links'] = [str(text).replace(punctuation, '') for text in df['links']]    
```

```python
# Lower text

df['summaries'] = [text.lower() for text in df['summaries']]
df['links'] = [text.lower() for text in df['links']]
```

```python
# Stopwords

stop_words = set(stopwords.words('english')) 

df['summaries_tokened'] = [word_tokenize(text) for text in df['summaries']]
df['links_tokened'] = [word_tokenize(text) for text in df['links']]

for i, word_tokens in enumerate(df['summaries_tokened']):
    text = [w for w in word_tokens if not w in stop_words]
    df['summaries'].iloc[i] = ' '.join(text)

for i, word_tokens in enumerate(df['links_tokened']):
    text = [w for w in word_tokens if not w in stop_words]
    df['links'].iloc[i] = ' '.join(text)
    
df = df.drop('summaries_tokened',axis = 1)   
df = df.drop('links_tokened',axis = 1)   
```

```python
# Remove digits 

for i, text in enumerate(df['summaries']):
    df['summaries'].iloc[i] = ''.join([word for word in text if not word.isdigit()])
    
for i, text in enumerate(df['links']):
    df['links'].iloc[i] = ''.join([word for word in text if not word.isdigit()])
```

```python
# Lemmatize

lemmatizer = WordNetLemmatizer()

for i, text in enumerate(df['summaries']):
    df['summaries'].iloc[i] = ' '.join([lemmatizer.lemmatize(word) for word in text.split(' ')])

for i, text in enumerate(df['links']):
    df['links'].iloc[i] = ' '.join([lemmatizer.lemmatize(word) for word in text.split(' ')])
```

```python
# Keep only nouns

for i, text in enumerate(df['summaries']):
    tokens = text.split()
    tags = nltk.pos_tag(tokens)
    
    df['summaries'].iloc[i] = [word for word, pos in tags if (pos == 'NN' or pos == 'NNP' or pos == 'NNS' or pos == 'NNPS')]
    
for i, text in enumerate(df['links']):
    tokens = text.split()
    tags = nltk.pos_tag(tokens)
    
    df['links'].iloc[i] = [word for word, pos in tags if (pos == 'NN' or pos == 'NNP' or pos == 'NNS' or pos == 'NNPS')]
    
```
