```python
import _pickle as cPickle

import pandas as pd
import numpy as np
```

```python
var = cPickle.load(open('../pkl/scraped_recipe_ingredients.pkl', 'rb'))
```

```python
var
```

```python
df_nyt = pd.read_csv('https://raw.githubusercontent.com/mtlynch/ingredient-phrase-tagger/master/nyt-ingredients-snapshot-2015.csv')
```

```python
ingr_list = df_nyt[['name']]
ingr_list
```

```python
ingr_list = list(ingr_list['name'])
ingr_list
```

```python
'flour' in var[0][0]
```

```python
n = 0
for i,recipe in enumerate(var):
    print(n)
    n += 1
    for j,ingredient in enumerate(recipe):
        for ingr in ingr_list:
            if str(ingr) in str(ingredient):
                var[i][j] = str(ingr)
```
